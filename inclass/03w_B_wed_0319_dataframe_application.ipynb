## df.query()

```python
df = pd.DataFrame({'A': range(1, 6),
                   'B': range(6, -3, -2),
                   'C': range(10, 5, -1),
                   'D clmns': range(3, -2, -1),
                   'E':['python', 'java', 'python','go', 'python']})
df
```

```python
df.query('A>B')
```

```python
df[df.A> df.B]
```

```python
df.query('C == 10')
```

```python
df.query('A>0')
```

```python
df.query('E== "python"')
```

```python
df.query("E== 'python'")
```

```python
df.query("E != 'python'")
```

```python
# df.query("A == D clmns") 에러
# 에러는 안나지만 불가능 df.query("A == 'D clmns'") 'A == "D clms"' 중간에 띄어쓰기가 있는 경우
df.query("A == `D clmns`") # 백팁키(TAB키 위) 를 사용하면 columns 이름으로도 가능
```

```python
df[ df.A == df["D clmns"]]
```

```python
df.query('A > 3 | B >8')
```

```python
df.query('(A > 3) | (B >8)')    # 위 데이터프레임과 달리 query에선 () 없이도 가능
```

```python
df.query('A > 2 & C <8')
```

```python
df.query( 'E in ["python", "go"]')
```

## 데이터프레임 수정

```python
import numpy as np
import pandas as pd

dates = pd.date_range("20250102", periods =6)
df = pd.DataFrame(np.arange(24).reshape(6, 4), index = dates)
df
```

```python
df.columns = list('ABCD')
df
```

```python
s1 = pd.Series([1, 2, 3, 4, 5], index = pd.date_range("20250102", periods = 5))
s1
```

```python
df["E"] = s1
df
```

## 값 수정

```python
df.loc['2025-01-02', 'A'] = 10
df
```

```python
df.loc[dates[1], 'B']= 50
df
```

```python
df.at[dates[2], df.columns[2]] = 100
df
```

```python
df.at[df.index[-1], df.columns[-1]] = 230
df
```

```python
df.iat[1, 2] = 60
df
```

```python
df.iat[3, 3] = 150
df
```
