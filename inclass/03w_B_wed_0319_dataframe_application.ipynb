```python
# 복사 및 열 추가
df2 = df.copy()
df2['재해자수'] = df[['사망(명)', '부상(명)']].sum(axis=1)
df2
```

```python
df2.sum()
```

```python
# 행 추가 loc location을 통해 만들어야 함
df2.loc['총계'] = df2.sum(axis =0)
df2
```
