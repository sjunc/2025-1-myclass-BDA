## 주요 칼럼 설명
| 컬럼명        | 데이터 타입         | 설명                                                                                  |
| ---------- | -------------- | ----------------------------------------------------------------------------------- |
| `date`     | `object` (문자열) | 해당 데이터가 수집된 날짜. 일반적으로 연-월 형식(예: `2000-01`)으로 되어 있습니다.                               |
| `pce`      | `float64`      | 개인 소비 지출(Personal Consumption Expenditures). 미국 내 소비자들의 소비 규모를 수치로 표현한 값입니다.        |
| `pop`      | `float64`      | 총 인구 수(Population)를 나타내는 지표로, 단위는 백만(Million)명입니다.                                  |
| `psavert`  | `float64`      | 개인 저축률(Personal Saving Rate). 개인의 가처분 소득 중 저축한 비율(%)을 의미합니다.                        |
| `uempmed`  | `float64`      | 실업 기간의 중앙값(Median Duration of Unemployment), 실업 상태가 지속된 기간의 중앙값을 주(week) 단위로 나타냅니다. |
| `unemploy` | `int64`        | 실업자 수(Unemployed Persons). 미국 내 실업자 수를 천(Thousand) 단위로 나타낸 정수 값입니다.                 |

## 정리하기
| Seaborn 함수      | 그래프 종류                    | 설명                                         |
| --------------- | ------------------------- | ------------------------------------------ |
| `lineplot()`    | 선 그래프 (Line Plot)         | 시계열 데이터나 연속형 변수의 추세를 시각화할 때 사용합니다.         |
| `scatterplot()` | 산점도 (Scatter Plot)        | 두 연속형 변수 간의 관계를 시각화할 때 사용합니다.              |
| `barplot()`     | 막대 그래프 (Bar Plot)         | 범주형 변수와 수치형 변수 간의 비교를 나타낼 때 사용합니다.         |
| `histplot()`    | 히스토그램 (Histogram)         | 단일 변수의 분포를 시각화할 때 사용합니다.                   |
| `boxplot()`     | 박스 플롯 (Box Plot)          | 변수의 분포, 중앙값, 이상치를 시각화할 때 유용합니다.            |
| `violinplot()`  | 바이올린 플롯 (Violin Plot)     | 박스플롯과 밀도그래프를 결합한 형태로, 분포를 보다 자세히 보여줍니다.    |
| `heatmap()`     | 히트맵 (Heatmap)             | 변수 간의 상관관계 또는 행렬 데이터를 시각화할 때 사용합니다.        |
| `pairplot()`    | 페어 플롯 (Pair Plot)         | 여러 변수 간의 관계를 한 번에 시각화할 수 있는 다변량 산점도 행렬입니다. |
| `regplot()`     | 회귀선 산점도 (Regression Plot) | 두 변수 간의 관계와 회귀선을 함께 나타낼 때 사용합니다.           |
| `countplot()`   | 카운트 플롯 (Count Plot)       | 범주형 변수의 빈도수를 시각화할 때 사용합니다.                 |

```python
from IPython.display import display_html

# 여러 데이터프레임 비교가 쉽게 옆쪽으로 표시한다.
def display_side_by_side(*args):      # * 여러 개의 인자 가 들어옴
    html_str = ''
    for df in args:
      html_str += df.to_html() + '&nbsp' * 4    # &nbsp 공란 4 칸
    display_html(html_str.replace('table', 'table style="display:inline"'), raw = True)
```

```python
import pandas as pd

# 중간고사 데이터
test1 = pd.DataFrame({'id'      : [1, 2, 3, 4, 5],
                      'midterm' : [60, 80, 70, 90, 85]})

# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id'    : [1, 2, 3, 6, 7],
                      'final' : [70, 83, 65, 95, 80]})
```

```python
test1
```

```python
test2
```

```python
display_side_by_side(test1, test2)    # 옆으로 나란히 데이터프레임 출력
```

```python
display_side_by_side(test1, test2, test1.merge(test2))
```
