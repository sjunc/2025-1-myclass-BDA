# 시리즈

```python
import numpy as np
import pandas as pd

s = ['희빈', '수영', '현수', '지호', '지민']
p = pd.Series([99, 84, 92, 65, 78], index = s)
p
```

std 표준 편차

```python
s = pd.Series([3, 1, 1, 2, None, np.nan, 4, 6, 6, np.nan])
s
```

```python
s.values
```

```python
s.count()
```

```python
s.value_counts()    # 가장 많은 값부터 세줌
```

```python
s.value_counts(dropna = True)
```

```python
s.value_counts(dropna = False)    # 결측값 포함
```

```python
s.value_counts(sort=False, dropna= False)
```

```python
s.value_counts(bins=4)    # 구간 정하기
```

```python
s.value_counts(bins=4, sort = False)   # ( = 불포함 [ = 포함
```

```python
s.value_counts(bins=[0.5, 3.5, 6.5], sort = False)    # 구간 직접 정해주기
```

# 날짜를 위한 함수 date_range()

```python
dates = pd.date_range("20250101", periods = 6)
dates
```

```python
dates = pd.date_range("20250101", end = '20250110')
dates
```

```python
dates = pd.date_range("20250101", periods = 6, freq = '3D')
dates
```

```python
dates = pd.date_range("20250101", periods = 6, freq = '3h')
dates
```

```python
dates = pd.date_range("20250101", periods = 6, freq='W') # 기본 일요일 시작
dates
```

```python
dates = pd.date_range("20250101", periods = 6, freq='W-MON')
dates
```

```python
dates = pd.date_range("20250101", periods = 6, freq='MS')   # MS 달 시작 ME 달 끝
dates
```

```python
dates = pd.date_range("20250301", periods = 6)
dates
```

```python
df = pd.DataFrame(np.random.rand(6, 4), index = dates,
                  columns = list('ABCD'))
df
```
