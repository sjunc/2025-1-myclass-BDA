```python
s = pd.Series([3, 1, 1, 2, None, np.nan, 4, 6, 6, np.nan])
s
```

```python
s.values
```

```python
s.count()
```

```python
s.value_counts()    # 가장 많은 값부터 세줌
```

```python
s.value_counts(dropna = True)
```

```python
s.value_counts(dropna = False)    # 결측값 포함
```

```python
s.value_counts(sort=False, dropna= False)
```

```python
s.value_counts(bins=4)    # 구간 정하기
```

```python
s.value_counts(bins=4, sort = False)   # ( = 불포함 [ = 포함
```

```python
s.value_counts(bins=[0.5, 3.5, 6.5], sort = False)    # 구간 직접 정해주기
```

## 데이터프레임 참조

```python
pf = pd.DataFrame(data =  [[25, 35, 8, 18],
                          [18, 27, 10, 20],
                          [17, 17, 10, 19],
                          [12, 22, 9, 20],
                          [22, 34, 8, 16]],
                  index = ['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
                  columns = ['중간', '기말', '과제', '출석'])
pf
```

```python
pf['과제']
```

```python
type(pf['과제'])      # type은 series
```

```python
pf[ ['과제'] ]    # [] 를 처리하면 type이 나오지 않는 데이터프레임 형태
```

```python
pf.중간
```

```python
pf.중간 > 20
```

```python
pf['중간'] > 20
```

```python
pf[pf['중간']> 20]          # 시험 문제
```

```python
pf[pf['출석']>= 19]
```
