## mpg 데이터 활용 groupby() agg()

```python
import pandas as pd

mpg = pd.read_csv('data/mpg.csv')
mpg.info()
```

```python
mpg.columns
```

```python
mpg.sample(frac = .05) # 5% sample
```

```python
# manufacturer와 drv는 둘다 index 다중 인덱스
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean'))    # 항목 2개이상 조회할 땐 항상 []으로 묶어줄 것
```

```python
mpg.groupby(['drv', 'manufacturer']).agg(mean_cty = ('cty', 'mean'))    # 항목 2개이상 조회할 땐 항상 []으로 묶어줄 것
```

```python
mpg.groupby(['manufacturer', 'category']).agg(mean_cty = ('cty', 'mean'))    # 항목 2개이상 조회할 땐 항상 []으로 묶어줄 것
```

```python
mpg.groupby(['category', 'manufacturer']).agg(mean_cty = ('cty', 'mean'), mean_hwy = ('hwy', 'mean'))
```

```python
#쿼리문

mpg.query('manufacturer == "audi"').groupby(['drv']).agg(n = ('drv', 'count'))
```

```python
mpg.query('category == "suv"') \
   .assign(total = (mpg['hwy'] + mpg['cty']) / 2) \
   .groupby('manufacturer') \
   .agg(mean_tot = ('total', 'mean')) \
   .sort_values('mean_tot', ascending = False) \
   .head(10)
mpg2 = mpg
mpg2.query("category=='suv'") \
    .assign(total = (mpg.hwy + mpg.cty)/2) \
    .groupby('manufacturer') \
    .agg(mean_total= ('total', 'mean')) \
    .sort_values('mean_total', ascending = False) \
    .head(10)
```

## 데이터 합치기

```python
# 중간
test1 = pd.DataFrame({'id': [1, 2, 3, 4, 5],
                      'midterm': [60, 80, 70, 90, 85]})
# 기말
test2 = pd.DataFrame({'id': [1, 2, 3, 4, 5],
                      'final': [70, 83, 65, 95, 80]})
```

```python
test1
```

```python
test2
```

```python
pd.merge(test1, test2)
```

```python
pd.merge(test1, test2, on = "id") # 기본
```

```python
pd.merge(test1, test2, on = 'id', how = 'inner') # inner 공통 사항만 합치기
```

```python
test1.merge(test2, on = 'id', how = 'inner') # method
```

```python
test2.merge(test1, on = 'id', how = 'inner') # 가능하지만 불편함
```

```python
# 학생 1~5번 시험 데이터 만들기
group_a = pd.DataFrame({'id'   : [1, 2, 3, 4, 5],
                        'test' : [60, 80, 70, 90, 85]})

# 학생 6~10번 시험 데이터 만들기
group_b = pd.DataFrame({'id'   : [6, 7, 8, 9, 10],
                        'test' : [70, 83, 65, 95, 80]})
```

```python
# concat 세로 합치기
pd.concat([group_a, group_b])
```

```python
pd.concat([group_a, group_b]).reset_index() # index 초기화 해서 0~9까지 원 인덱스는 열로
```

```python
pd.concat([group_a, group_b]).reset_index(drop = True) # 원 index 제거
```

```python
import pandas as pd

df1 = pd.DataFrame([[1, 'apple'], [2, 'banana']],
                  columns = ['id', 'fruits'])
df1
```

```python
df2 = pd.DataFrame([[1, 3000], [5, 500]],
                  columns = ['id', 'price'])
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on = 'id')
```

```python
pd.merge(df1, df2, on = 'id', how = 'inner')
```

```python
pd.merge(df1, df2, on = 'id', how = 'left') # how = 'left' 왼쪽만 다 가지고 감
```

```python
pd.merge(df1, df2, on = 'id', how = 'right') # how = 'right' 오른쪽만 다 가지고 옴. 없는 값은 Nan 값
```
