## 7번

```python
df1 = pd.DataFrame({'sub':['AI', 'Python', 'Java'],
                    'score':[80, 77, 99]})
df2 = pd.DataFrame({'sub':['AI', 'Python', 'C'],
                    'book':[30000, 28000, 25000]})
print(df1)
print(df2)
```

### merge 문제
1. python 정답 : 3 shape는 []안에 숫자에 따라 축 개수를 묻는 부분
2. 28000  정답
3. Java   정답
4. 3      정답 : 9 cross는 전부 곱해서 9 X 4의 테이블을 만듦

```python
df1
```

```python
df1.shape  # (행, 렬) 반환
df1.shape[1] # [1] 열 개수 반환
df1.shape[0] # [0] 행 개수 반환
```

```python
# 7-1 3인 이유 기본 merge는 항목 4개 그 중 열의 개수를 묻고 있음.
df1.merge(df2).shape[1]
```

```python
df1.merge(df2)
```

```python
# how는 뭘 남길지 좌측을 남기는 기준으로 merge
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'left')['book'][1]
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'outer').iloc[2, 0]
```

```python
df1.merge(df2, how = 'cross') #모든 행끼리 조합
```

```python
len(df1.merge(df2, how = 'cross'))
```
