```python
import numpy as np
import pandas as pd

df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, np.nan],
                   [np.nan, 3, np.nan, 4]],
                  columns=list('ABCD'))
df
```

```python
df.fillna(10)
```

```python
df
```

```python
df.fillna('missing', axis = 0)
```

```python
df.fillna('missing', axis = 1)    # 기본적으로 axis =0 과 차이가 없음.
```

```python
df.fillna('missing', limit = 1)  # 각 열에 대해 처음만 바뀜
```

```python
df.fillna('missing', limit = 2)
```

```python
df.fillna('missing', limit = 2, axis = 0) # 기본이 0
```

```python
df.fillna('missing', limit = 1, axis =1)  # 세로
```

```python
df
```

```python
values = {'A': 100, 'B': 200, 'C': 300, 'D': 400}
df.fillna(values)
```

```python
values = {'A': 100, 'B': 200, 'C': 300, 'D': 400}
df.fillna(values, limit = 2)
```

```python
df    # df는 변경되지 않음
```

```python
s = pd.Series([100, 200, 400], index = list('ACD'))
s
```

```python
df.fillna(s, limit = 1)
```

```python
df2 = pd.DataFrame(-np.ones((4,4)), columns = list("ABCE"))
df2
```

```python
df.fillna(df2)
```

```python
df
```

```python
df.fillna(method = 'ffill')  # front of 앞쪽에 있는 값으로 채우기 오류: 앞으로 없어질 것임
```

```python
df.ffill()    # ffill 함수로 변경됨
```

```python
df.ffill(limit = 1)
```

```python
df.ffill(limit = 1, axis =1)
```

```python
# df.fillna(method = 'bfill', limit = 1) # back of 뒤에 있는 값으로 채우기 오류
df.bfill(limit = 1)
```

```python
df.bfill(limit = 1, axis=1)
```

```python
df
```

```python
df.mean()   #nan 값은 제거하고 평균을 구함
```

```python
df.fillna(df.mean())
```

```python
df.mean()['A':'B']
```

```python
df.fillna(df.mean()['A':'B'])
```

```python
df.fillna(df.mean()['D'])
```

```python
df.fillna(df.mean()['D'], limit =1)
```

```python
df.fillna(df.mean()['D'], limit =1, axis =1)
```

### 타이타닉 열(변수) 설명
- survived : 0 = 사망, 1 = 생존
- pclass : 1 = 1등석, 2 = 2등석, 3 = 3등석
- sex : male = 남성, female = 여성
- age : 나이
- sibsp : 타이타닉 호에 동승한 자매 / 배우자의 수
- parch : 타이타닉 호에 동승한 부모 / 자식의 수
- fare : 티켓 요금
- embarked : 탑승지, C = 셰르부르, Q = 퀸즈타운, S = 사우샘프턴
- class : First = 1등석, Second = 2등석, Third = 3등석
- who : 남/여/아이, 'man', 'woman', 'child'
- adult_male : 남자 어른, True/False
- deck : 방 위치, 'A', 'B', 'C', 'D', 'E', 'F', 'G', NaN
- embark_town : 탑승지
- alive : 생존, yes/no
- alone : 1인 탑승, True/False
